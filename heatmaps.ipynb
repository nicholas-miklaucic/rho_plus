```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import rho_plus as rp

theme, cs = rp.mpl_setup(False)
```

```python
def teeth_plot(palette):
    plt.imshow(plt.imread('docs/source/colourmaptest.tif'), cmap=palette)
    plt.gca().set_axis_off()

teeth_plot('jet')
```

```python
glue = sns.load_dataset("glue").pivot(index="Model", columns="Task", values="Score")
```

```python
from rho_plus.oklch_palettes import OklchPalette, oklch2hex, hex2oklch

hex2oklch('#' + theme['axes.facecolor'])
```

```python
theme, cs = rp.mpl_setup(True)
cmap = rp.oklch_palettes.OklchPalette(
    'aquaria',
    [
        (0.23, 0.05, 315),
        (0.9, 0.1, 155),
        (0.95, 0.15, 140),
    ],
    (False,False), correct='cam16')

teeth_plot(cmap.as_mpl_cmap())
plt.show()

plt.show()

sns.heatmap(glue, annot=True, cmap=cmap.as_mpl_cmap(), fmt='.0f', square=True)
```

```python
rp.mpl_setup(False)

viridia = OklchPalette("viridia", [(0.34, 0.13, 300), (0.57, 0.08, 200), (0.8, 0.13, 100)], [False, False])

teeth_plot(viridia.as_mpl_cmap())
plt.show()

pg = sns.load_dataset('penguins')
def scatterplot(palette):
    plt.scatter(data=pg, x='bill_length_mm', y='bill_depth_mm', c='body_mass_g', cmap=palette)
    plt.xlabel('Bill Length (mm)')
    plt.ylabel('Bill Depth\n(mm)', rotation='horizontal', ha='right')
    plt.title('Penguin Data')
    clb = plt.colorbar()
    clb.ax.set_title('Body Mass (g)')

scatterplot(viridia.rev().as_mpl_cmap())
```
